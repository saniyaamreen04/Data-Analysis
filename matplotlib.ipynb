### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[4,7,10,12,15,17]
y=[6,9,3,10,15,16]
```

```python
# plot function
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Basic Plotting");
```

```python
plt.plot(x,y,'r*');   # matplotlib markers
```

```python
plt.plot(x,y,color='red',marker='s',linewidth=4,linestyle='--',markersize=9,markerfacecolor='green',markeredgewidth=3,markeredgecolor='black');
```

```python
# Multiple plotting
# subplot function
# plt.subplot(nrows,ncols,plot number)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Plot 1")

plt.subplot(1,2,2)
plt.plot(x,y,'h')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Plot 2")

plt.tight_layout();  # to avoid overlapping
```

```python
label=['ram','shiva','giri']
score=[89,90,78]
```

```python
plt.plot(label,score)
```

```python
plt.pie(score,labels=label,autopct='%1.0f%%');
```

```python
plt.figure(figsize=(10,10))  #width , height
plt.subplot(1,2,1)   # single row,two colums, one plot 
plt.bar(label,score)

plt.subplot(1,2,2)
plt.pie(score,labels=label,autopct='%0.1f%%');
```
