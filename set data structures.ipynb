## Set data structure

Set is a collection of data(Multiple values), unordered, unindexed.

Format of using the set is { } curly brackets.

```python
# creating/defining the set
colors={'Blue','brown','blue','green','white','yellow','orange'}
print(colors)   # unordered
```

```python
colors[0]
```

```python
# add function
colors.add('violet')
print(colors)
```

```python
# remove function,
colors.remove('white')
print(colors)
```

```python
# pop function
colors.pop()  # without index
print(colors)
```

```python
del colors   # whole data will be deleted completely
```

```python
print(colors)
```

### Dictionary data structure

Dictionary is a collection of data(Multiple values), changeable format, indexed.

Format of using the dictionary is {key/index:value} curly brackets

```python
# creating/defining the dictionary
cars={'brand':['Ford','BMW','Maruti'],'Model':['Endevour','X7','Swift'],'colors':['black','blue','silver']}
print(cars)
```

```python
cars['brand']
```

```python
cars['brand'][1]
```

```python
cars['colors']
```

```python
cars['colors'][0]='grey'
print(cars)
```

```python
# popping out with index
cars.pop('colors')
print(cars)
```

```python
# deleting the particular value
del cars['Model']
print(cars)
```

```python
del cars  # whole data will be deleted
```

```python
print(cars)
```

### Functions

Functions are a block of code(Multiple statements), it just creates, and returns the output until the function name is called.

```python
# Simple function
def fruit():
    print("Apple is good for health")
    print("It is red in color")
```

```python
fruit()
```

```python
# single argument
def red(x):
    print(x%45)
    print(x//30)
    print(x**2)
```

```python
red(100)
```

```python
# Multiple argument
def add(x,y):
    print(x%y)
    print(x//y)
    print(x**y)
    print(x<y)
```

```python
add(23,67)
```

```python
def code(name,language):
    print(name+" is coding in "+language)
```

```python
code('Pavan','Python')
```

```python
code('Python','Giri')
# order of passing the values to the arguements matters
# unexpected result will be formed
```

```python
# argument=value
code(language='Python',name='Giri')
```
